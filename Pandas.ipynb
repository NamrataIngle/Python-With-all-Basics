Aggregation function in pandas aggregation - Computing a summary stastic about each group

```python
# using agg() function you can go with many functions at same time

df[['phd','salary']].agg(['min','mean','max','mad','std'])          #mad - mean absolute deviation
```

```python
print(df[['salary']].agg('skew'))      #skewness
print(df[['salary']].agg('kurt'))       #kurtosis
```
